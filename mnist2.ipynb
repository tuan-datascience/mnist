```python
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
from skimage.io import imshow
from tensorflow.contrib.layers import flatten,fully_connected,conv2d,max_pool2d

rows = 28
cols = 28
num_features = rows*cols
num_cates = 10

# Build Model
# Placeholder for data
X = tf.placeholder(dtype=tf.float32,shape=(None,rows,cols,1), name='InputData')
y = tf.placeholder(dtype=tf.int32,shape=(None),name='Labels')


with tf.name_scope('CNN'):
    conv_lay1 = conv2d(X,32,[5,5],stride=1,padding='SAME')
    conv_lay1 = max_pool2d(conv_lay1,[2,2])
    
    conv_lay2 = conv2d(conv_lay1,16,[5,5],stride=1,padding='SAME')
    conv_lay2 = max_pool2d(conv_lay2,[2,2])
    
    conv_lay3 = conv2d(conv_lay2,8,[5,5],stride=1,padding='SAME')
    conv_lay3 = max_pool2d(conv_lay3,[2,2])
    
    flat = flatten(conv_lay3)

    hidden1=fully_connected(flat,128,scope="hidden1")
    hidden2=fully_connected(hidden1,64,scope="hidden2")
    logits=fully_connected(hidden2,10,scope="outputs",activation_fn=None)
with tf.name_scope("Loss"):
    entropy=tf.nn.sparse_softmax_cross_entropy_with_logits(labels=y,logits=logits)
    loss=tf.reduce_mean(entropy, name="loss")
with tf.name_scope("Train"):
    optimizer=tf.train.GradientDescentOptimizer(learning_rate=0.01)
    train_op=optimizer.minimize(loss)
with tf.name_scope("Eval"):
    correct=tf.nn.in_top_k(logits,y,1)
    accuracy=tf.reduce_mean(tf.cast(correct,tf.float32))

from tensorflow.examples.tutorials.mnist import input_data
mnist=input_data.read_data_sets("/tmp/data/")
n_epochs=100
batch_size=50

with tf.Session() as sess:
    tf.global_variables_initializer().run()
    for epoch in range(n_epochs):
        for i in range(mnist.train.num_examples//batch_size):
            X_batch,y_batch=mnist.train.next_batch(batch_size)
            X_batch = X_batch.reshape(X_batch.shape[0],rows,cols,1)
            sess.run(train_op,feed_dict={X:X_batch,y:y_batch})
        acc_train = accuracy.eval(feed_dict={X:X_batch,y:y_batch})
        acc_test = accuracy.eval(feed_dict={X:mnist.test.images.reshape(mnist.test.images.shape[0],rows,cols,1),
                                           y:mnist.test.labels})
        print(epoch,"Train acc: ", acc_train,"Test acc: ", acc_test)
    saver=tf.train.Saver().save(sess,"./my_model1.ckpt")


```
